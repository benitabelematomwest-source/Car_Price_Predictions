```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
dummies = pd.get_dummies(df["Car Model"], drop_first=True)
dummies.head()
```

```python
df2 = pd.concat([df.drop("Car Model", axis=1), dummies], axis=1)
df2.head()
```

```python
for col in df2.columns:
    print(col)
```

```python
X = df2.drop("Sell Price($)", axis=1)
y = df2["Sell Price($)"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
score = model.score(X, y)
print("Model Score (Accuracy):", score)

```

```python
def predict_price(age, mileage, bmw_x5=0, audi_a5=0):
    return model.predict([[mileage, age, bmw_x5, audi_a5]])[0]
```

```python
# 1.Predict price of Mercedes Benz, 4 yrs old, mileage 45000
q1 = predict_price(4, 45000, 0, 0)
q1
```
